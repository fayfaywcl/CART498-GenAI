```python
import torch

device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")

from diffusers import StableDiffusionPipeline


pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=torch.float16,
    variant="fp16",
    ).to(device)


prompt1 = "A surreal collage image that include a top-down map floating in the sky, giant elephant pointing at buildings made of stacked coins, tiny ants crawling over coins and operating miniature cranes, dark smoke forming abstract shapes in the air, floating city blocks collaged together, dreamlike, whimsical"
pipe(prompt1).images[0]

prompt2 = "A surreal collage image that include a City framed like a vintage poster, smoke-shaped skyscrapers inhabited by tiny ants, giant elephant sitting on a cloud-like chair counting floating coins and fragmented map pieces surrounding the scene, ironic and playful, whimsical"
pipe(prompt2).images[0]

prompt3 = "A surreal collage image that a cityscape with the giant elephant stepping over a factory line of many big ants on industrial tracks which holding big money coins, buildings made of smoke and twisted forms, one ant lying on the elephant’s step, clouds and smoke forming abstract shapes,background do not be too dark, whimsical ,in blue and grey tone."
pipe(prompt3).images[0]

# Generate and save the image
#image 1
image1 = pipe(prompt1).images[0]
image1.save("surreal_collage_1.png")

image2 = pipe(prompt2).images[0]
image2.save("surreal_collage_2.png")

image3 = pipe(prompt3).images[0]
image3.save("surreal_collage_3.png")
```
