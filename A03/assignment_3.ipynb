```python
# Access secret keys via
from google.colab import userdata
OPENAI_API_KEY = userdata.get('MyKey')

# Import OpenAI API and set up the key
from openai import OpenAI
client = OpenAI(api_key=(OPENAI_API_KEY))

import re

# getting AI result
def gpt_multiply(a, b):
    prompt = f"Multiply {a} * {b}. Return ONLY the final number."

    response = client.chat.completions.create(
        model="gpt-4.1",
        messages=[{"role": "user", "content": prompt}],
        temperature=0
    )

    text = response.choices[0].message.content
    numbers = re.findall(r"-?\d+", text)
    return int(numbers[0]) if numbers else None
```

```python

#printing message (Self-deprecation mechanism)
def self_deprecating_message(errors):
    prompt = f"""
    You have made {errors} math mistakes so far.
     Respond with a short, humorous, self-deprecating message.
    If errors > 2, show clear frustration and defeat.
    Treat math like a hardest course you are tired of taking .
    """

    response = client.chat.completions.create(
        model="gpt-4.1-mini",
        top_p = 0.8,
        temperature=1.8,

        messages=[{
            "role": "developer",
            "content": """
                You are an inner monologue.
                Tone: aggressive frustration mixed with affectionate university student humor, will get frustrated or defeated over multiple errors.
                Style: self-blaming, dramatic, sarcastic.
                Metaphors: math is the hardest course that refuses to make sense.
                Length: 3 sentences.
            """
        },
        {
            "role": "assistant",
            "content": "I’ve made 4 mistakes and this math class is clearly grading me personally. I’ve tried notes, calculators, and caffeine—now I’m just ready to drop out and cry in the library.At this point, there’s no way I’m ever graduating."
        },
        {
            "role": "user",
            "content": prompt
        }]
    )

    return response.choices[0].message.content
```

```python

# Main Function

def worst_mathematician_bot(n, i):
    global errors
    current = n
    errors = 0

    print(f"\nStarting base = {n}, iterations = {i}")

    for step in range(i):
        #python code for calculations
        correct = current * current
        gpt_result = gpt_multiply(current, current)

        print(f"\nStep {step+1}: {current} * {current}")
        print("Correct result:", correct)
        print("GPT result:", gpt_result)

        if gpt_result != correct:
            errors += 1
            print(self_deprecating_message(errors))

        current = correct
```

```python
worst_mathematician_bot(8, 9)
worst_mathematician_bot(16, 9)
worst_mathematician_bot(9, 7)
```
